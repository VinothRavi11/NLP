```python
pattern = r'\s'
data['twitts'].apply(lambda x: re.sub(pattern, '', x))
```

```python
pattern = r'\s'
data['char_count']= data['twitts'].apply(lambda x: len(re.sub(pattern, '', x)))
```

```python
data['word_count']= data['twitts'].apply(lambda x: len(x.split()))
```

```python
data
```

```python
data['avg_word_length']=data['char_count']/data['word_count']
```

```python
data
```

```python
data['hash_tags_count']=data['twitts'].apply(lambda x: len(re.findall(r'#\w+', x)))
data
```

```python
data['mention_count']=data['twitts'].apply(lambda x: len(re.findall(r'@\w+', x)))
data
```

```python
data['twitts'].apply(lambda x: len([word for word in x.split() if word.isupper()]))
```

```python
data['twitts'].apply(lambda x: re.sub(r'\s+','',x).lower())
```

```python
pattern = r'@\w+'
data['twitts'].apply(lambda x: re.sub(pattern, '', x))
```

```python
pattern = r'[^\w\s]'
data['twitts'].apply(lambda x: re.sub(pattern, '', x))
```

```python
pattern= r'\w+@\w+.\w+'
```

```python
data['email'] = data['twitts'].apply(lambda x: re.findall(pattern=pattern, string=x))
data['email'].sample(5)
```

```python
pattern = r'\b[A-Za-z0-9._%+-]+@[A-ZA-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
data['email']=data['twitts'].apply(lambda x: re.findall(pattern = pattern, string = x))
data['email'].sample(5)
```

```python
pattern  = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
```

```python
data['email']=data['twitts'].apply(lambda x: re.findall(pattern = pattern, string = x))
data['email'].sample(5)
```
